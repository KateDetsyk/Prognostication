**show the positive and negative relationships of the predictors with the target variable**

The SHAP value plot can further show the positive and negative relationships of the predictors with the target variable. 

This plot consists of dots and the meaning of the dots :
* Vertical location of the dot shows what feature it is.
* Color of the dot shows whether that feature was high or low for that row of the dataset.
* Horizontal location of the dots shows whether the effect of that value caused a higher or lower prediction.

Let's look on `distance to the nearest MRT station`. If it become bigger - than price(target Y) become smaller 
and vice versa if `distance to the nearest MRT station` smaller that price higher (people want to live closer to the stations).

Same with the `house age`, the older is house - the lower is price, because customers don't want to pay a lot for the old house if thay can buy new.

With `longitude`, `latitude` and `transaction date` another situation, the bigget these values - the higher price and vice versa.

```python
def ABS_SHAP(df_shap, df):
    
    # Make a copy of the input data as pd.DataFrame
    shap_v = pd.DataFrame(df_shap)
    feature_list = df.columns
    shap_v.columns = feature_list
    df_v = df.copy().reset_index().drop('index', axis=1)
    
    # Determine the correlation in order to plot with different colors
    corr_list = list()
    for i in feature_list:
        b = np.corrcoef(shap_v[i], df_v[i])[1][0]
        corr_list.append(b)
    corr_df = pd.concat([pd.Series(feature_list), pd.Series(corr_list)],axis=1).fillna(0)
    # Make a data frame. Column 1 is the feature, and Column 2 is the correlation coefficient
    corr_df.columns  = ['Feature', 'Corr']
    corr_df['Sign'] = np.where(corr_df['Corr'] > 0, '#ff0d57', '#1e88e5')
    
    # Plot it
    shap_abs = np.abs(shap_v)
    k=pd.DataFrame(shap_abs.mean()).reset_index()
    k.columns = ['Feature', 'SHAP_abs']
    k2 = k.merge(corr_df, left_on = 'Feature', right_on='Feature', how='inner')
    k2 = k2.sort_values(by='SHAP_abs', ascending = True)
    colorlist = k2['Sign']
    ax = k2.plot.barh(x='Feature', y='SHAP_abs', color = colorlist, figsize=(8, 6), legend=False)
    ax.set_xlabel("SHAP Value (Red = Positive Impact, Blue = Negative Impact)")
    
ABS_SHAP(shap_values, X_train)
```

```python
def print_image(index, arr):
    temp = arr[index]
    temp = temp.reshape(28,28).astype('uint8')
    plt.imshow(temp)
    
print_image(3, x_train)
```

**prepare data**

```python
DIM = 28 # image dimensions
x_train = x_train.reshape(x_train.shape[0], DIM, DIM, 1)
x_test = x_test.reshape(x_test.shape[0], DIM, DIM, 1)
input_shape = (DIM, DIM, 1)

# uint to float
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

print("Train shape = ", x_train.shape)
print("Test shape = ", x_test.shape)
```

**train model**

```python
y_train = keras.utils.to_categorical(y_train, 10) # to binary matrices
y_test = keras.utils.to_categorical(y_test, 10)

# define model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=128,
                    epochs=10,
                    verbose=1,
                    validation_data=(x_test, y_test))
```

```python
sns.lineplot(range(1, 11), history.history['loss'])
sns.lineplot(range(1, 11), history.history['val_loss'])
plt.xlabel('Epoch')
plt.xlabel('Loss')
plt.legend(['train', 'validation'])
plt.title('Learning curve');
```
